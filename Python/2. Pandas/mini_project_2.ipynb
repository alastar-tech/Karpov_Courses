```python
bookings.columns = map(str.lower, bookings.columns)
bookings.columns = bookings.columns.str.replace(' ', '_')
bookings.info()
```

```python
bookings_success = bookings.query("is_canceled == 0")
bookings_success.groupby('country') \
                .agg({'hotel': 'count'}) \
                .sort_values('hotel', ascending = False) \
                .head(5)
```

```python
arr_2016 = bookings.query("arrival_date_year in [2015, 2016, 2017] and is_canceled == 1") \
            .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
            .agg({'hotel': 'count'}) \
            .sort_values('hotel', ascending=False)
```
