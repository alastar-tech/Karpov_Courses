```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# найти все произведения, связанные с подсолнухами (id-шники)

r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?q=sunflowers')
res = r.json()

sunflower_ids = res['objectIDs']
```

```python
#запрос по объекту с id = 437980
r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/649813')
obj_test = r_test.json()

obj_test
```

```python
#создаем пустой ДФ
main_df = pd.DataFrame(data={'objectID': [], 
                            'title': [],
                            'artistDisplayName': [],
                            'department': [],
                            'objectBeginDate': [],
                            'objectEndDate': [],
                            'period': [],
                            'objectName': [],
                            'culture': []})

#создаем список ключей, которые надо вытащить из поулченного через API запроса
columns = ['objectID', 
           'title', 
           'artistDisplayName', 
           'department',
           'objectBeginDate',
           'objectEndDate',
           'period',
            'objectName',
            'culture']
```

```python
for i in sunflower_ids:
    a = []
    r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{i}') #запрос к сайту
    obj_test = r_test.json()            #получение словаря с данными        


    try:                                   
        for col in columns:                  #цикл доавления значений нужных полей(из columns) в список*
            a.append(obj_test[col])
        main_df.loc[len(main_df.index)] = a  #добавление *списка к концу ДФ
    except:
        continue
```
