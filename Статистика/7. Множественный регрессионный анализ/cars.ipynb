```python
import pandas as pd
```

```python
cars = pd.read_csv('cars.csv')
```

```python
cars.head()
```

```python
#проверим дф на пропущенные значения и типы данных
cars.info()
```

```python
#сгенерируем первый признак
#выделим из поля CarName марку автомобиля (разделим по пробелу и возьмем первыое значение)

cars['brand'] = cars.CarName.apply(lambda x: x.split(' ')[0])
cars = cars.drop(columns=['CarName','car_ID']) #удалим лишние столбцы
```

```python
#посчитаем кол-во уникальных марок в дф
cars.brand.nunique()
```

```python
cars.brand.unique()
```

```python
#заменим дубликаты марок с ошибками

cars.loc[cars['brand'] == 'maxda', 'brand'] = 'mazda'
cars.loc[cars['brand'] == 'Nissan', 'brand'] = 'nissan'
cars.loc[cars['brand'] == 'porcshce', 'brand'] = 'porsche'
cars.loc[cars['brand'] == 'toyouta', 'brand'] = 'toyota'
cars.loc[cars['brand'] == 'vokswagen', 'brand'] = 'volkswagen'
cars.loc[cars['brand'] == 'vw', 'brand'] = 'volkswagen'
```

```python
#кол-во марок после очистки
cars.brand.nunique()
```

```python
#для построение регрессионной модели оставим часть предикторов (независимых переменных)

cars = cars[['brand', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower','price']]

#найдем корелляцию между переменными
cars.corr()
```

```python
cars.info()
```

```python
#т.к. линейная регрессия не работает с категориальными ПН, проведем One-Hot Encoding. 
#Заменим категориальные ПН на фиктивные числовые

#отдельно для строковых переменных

df_dummy = pd.get_dummies(data=cars[['brand',\
                                     'fueltype',\
                                     'aspiration',\
                                     'carbody',\
                                     'drivewheel'\
                                     ,'enginetype'\
                                     ,'cylindernumber']], drop_first = True)
df_dummy
```

```python
#One-Hot Encoding для всей таблицы

df_dummy_total = pd.get_dummies(data=cars[['brand'\
                                     , 'fueltype'\
                                     , 'aspiration'\
                                     , 'carbody'\
                                     , 'drivewheel'\
                                     , 'wheelbase'\
                                     , 'carlength'\
                                     , 'carwidth'\
                                     , 'curbweight'\
                                     , 'enginetype'\
                                     , 'cylindernumber'\
                                     , 'enginesize'\
                                     , 'boreratio'\
                                     , 'horsepower'\
                                     , 'price']], drop_first = True)
df_dummy_total
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf 

# способ первый
# Y = одномерный массив с ЗП, X - массив со всеми нужными нам НП

X = sm.add_constant(cars.horsepower)  # добавить константу, чтобы был свободный член
Y = cars.price
model = sm.OLS(Y, X)  # говорим модели, что у нас ЗП, а что НП
results = model.fit()  # строим регрессионную прямую
print(results.summary())  # смотрим результат

# способ второй, потенциально более удобный

#results = smf.ols('Y ~ X1 + X2 + ... + Xn', cars).fit()
#print(results.summary())
```
